```python
 import cv2
```

```python
 face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml') # We 
 eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml') # We load the cascade
```

```python
def detect(gray, frame): # We create a function that takes as input the image in 
  faces = face_cascade.detectMultiScale(gray, 1.3, 5) # We apply the detectMul 
  for (x, y, w, h) in faces: # For each detected face:
    cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2) # We paint a re 
    roi_gray = gray[y:y+h, x:x+w] # We get the region of interest in the bla 
    roi_color = frame[y:y+h, x:x+w] # We get the region of interest in the c 
    eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 3) # We apply the det 
    for (ex, ey, ew, eh) in eyes: # For each detected eye:
      cv2.rectangle(roi_color,(ex, ey),(ex+ew, ey+eh), (0, 255, 0), 2) # W 
    return frame # We return the image with the detector rectangles.
```

```python
video_capture = cv2.VideoCapture(0,cv2.CAP_V4L) # We turn the webcam on.
```

```python
while True: # We repeat infinitely (until break):
  _, frame = video_capture.read() # We get the last frame.
  gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # We do some colour transform 
  canvas = detect(gray, frame) # We get the output of our detect function. 
  cv2.imshow('Video', canvas) # We display the outputs.
  if cv2.waitKey(1) & 0xFF == ord('q'): # If we type on the keyboard:
    break # We stop the loop.
video_capture.release() # We turn the webcam off.
cv2.destroyAllWindows() # We destroy all the windows inside which the images wer
```
